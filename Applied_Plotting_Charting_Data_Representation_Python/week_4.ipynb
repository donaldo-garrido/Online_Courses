### Code

In the next section I write the code used to generate the images presented to demonstrate the research question.

```python
# Import libraries
import pandas as pd
```

```python
# Set the rows and columns to use
rows_to_skip = list(range(6))
colums_to_use = [0,1,14]

# SBI stands for Social Backwardness Index
headers = ['ID', 'State', 'SBI']

def concat_all(prefix='df_'):
    dfs = [df for name, df in globals().items() if name.startswith(prefix)
           and isinstance(df, pd.DataFrame)]
    return pd.concat(dfs, axis=0)

# Read the files for available years (2000, 2005, 2010, 2015, 2020)
df_2000 = pd.read_csv('IRS_ent_mun_2000_2020/IRS_entidades_mpios_2000.csv',usecols=colums_to_use, 
                        skiprows=rows_to_skip, names=headers)#, index_col='ID')
df_2005 = pd.read_csv('IRS_ent_mun_2000_2020/IRS_entidades_mpios_2005.csv',usecols=colums_to_use, 
                        skiprows=rows_to_skip, names=headers)#, index_col='ID')
df_2010 = pd.read_csv('IRS_ent_mun_2000_2020/IRS_entidades_mpios_2010.csv',usecols=colums_to_use, 
                        skiprows=rows_to_skip, names=headers)#, index_col='ID')
df_2015 = pd.read_csv('IRS_ent_mun_2000_2020/IRS_entidades_mpios_2015.csv',usecols=colums_to_use, 
                        skiprows=rows_to_skip, names=headers)#, index_col='ID')
df_2020 = pd.read_csv('IRS_ent_mun_2000_2020/IRS_entidades_mpios_2020.csv',usecols=colums_to_use, 
                        skiprows=rows_to_skip, names=headers)#, index_col='ID')

# Concatenate all DataFrames
df = concat_all()
```

```python
# Get the data to plot into DataFrame of 8 states taken randomly

years = [2000,2005,2010,2015,2020]
states_evolution = pd.DataFrame({'Aguascalientes':df[df['ID'] == 1]['SBI'].values,
                         'Coahuila':df[df['ID'] == 5]['SBI'].values,
                         'Chiapas':df[df['ID'] == 7]['SBI'].values,
                         'Guerrero':df[df['ID'] == 12]['SBI'].values,
                         'Morelos':df[df['ID'] == 17]['SBI'].values,
                         'Oaxaca':df[df['ID'] == 20]['SBI'].values,
                         'Sinaloa':df[df['ID'] == 25]['SBI'].values,
                         'Zacatecas':df[df['ID'] == 32]['SBI'].values}, index=years)
```

```python
states_evolution
```

### Image

```python
# Plot the DataFrame
states_evolution.plot(title='Social Backwardness for five states of Mexico', 
              xticks=years, figsize=(8,9)).set(ylabel='Social Backwardness Index', xlabel='Year');
```

### Discussion

The image is a graph of the evolution of the index of social backwardness for the states of Aguascalientes, Coahuila, Chiapas, Guerrero, Morelos, Oaxaca, Sinaloa and Zacatecas between the period from 2000 to 2020, taking the data every 5 years as it is given in the databases.

As can be seen, the states with the greatest social backwardness show an upward trend in the index, while those with a moderate lag or no lag (negative) compared to the national average remain at similar levels, if not slightly better. This shows the trend of the states with the highest and lowest backwardness indexes.
